```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import matplotlib.pyplot as plt
```

```python
true = pd.read_csv('D:/Download/test 2.0/Match 37541/true.csv',delimiter=",")
predicted = pd.read_csv('D:/Download/test 2.0/Match 37541/predicted.csv',delimiter=",")

true = list(true['coup'])
predicted = list(predicted['coup'])
```

```python
labels = ['Net','Lob','Smash','Drop','Clear']

conf_matrix = confusion_matrix(y_true=true, y_pred=predicted,labels=labels)
#
# Print the confusion matrix using Matplotlib
#
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')

ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
print("Precision Score : ",precision_score(true , predicted, 
                                           average='micro'))
print("Recall Score : ",recall_score(true , predicted,
                                           average='micro'))
```
