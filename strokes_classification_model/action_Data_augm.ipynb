```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL
import tensorflow as tf
import pathlib
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_dir = pathlib.Path("D:/EFREI/project/data/dataset/strokes/test 1/Lob-Net")
data_dir
```

```python
Strokes = list(data_dir.glob('*.jpg'))
```

```python
image_count = len(Strokes)
print(image_count)
```

```python
X = []
for image in Strokes:

    img = cv2.imread(str(image))
    resized_img = cv2.resize(img,(180,180))
    X.append(resized_img)

X = np.array(X)
```

```python
data_augmentation = tf.keras.Sequential([
tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal'),
tf.keras.layers.experimental.preprocessing.RandomZoom(0.1)
])
```

```python
plt.axis('off')
plt.imshow(X[0])
```

```python
image_count = 1530
for i in range(len(Strokes)):

    img1 = data_augmentation(X)[i].numpy().astype("uint8")
    cv2.imwrite("./data/dataset/strokes/test 1/Lob-Net/"+str(image_count)+".jpg",img1)
    image_count+=1
```
